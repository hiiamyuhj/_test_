```python
import pandas as pd

df = pd.read_csv('world.csv')
df
```

```python
df1=df.query('國家 == "台灣"')
df1
```

```python
df2 = df1.sort_values(by='新增確診數',ascending=False)
df2.head(5)
```

```python
df3=df2.set_index('日期')
df3
```

```python
df3['no_rank'] = df3['新增確診數'].rank(method='first',ascending=False).astype(int)
df3
```
